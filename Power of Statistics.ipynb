## Conclusion 

The describe() function is also useful because it reveals a variety of key stats all at once. Python also has separate functions for the mean, median, standard deviation, minimum, and maximum. Earlier in the program, you used mean() and median() to detect outliers. These individual functions are also useful if you want to do further computations based on descriptive stats. 

### Example 

```python
#### Measures of Central Tendency
mean_literacy_rate = education_districtwise['OVERALL_LI'].mean()
print("Mean Literacy Rate:", mean_literacy_rate)

median_literacy_rate = education_districtwise['OVERALL_LI'].median()
print("Median Literacy Rate:", median_literacy_rate)

BLOCKS = education_districtwise['BLOCKS'].mode()
print("Mode BLOCKS:", BLOCKS	)
```

```python
#### Measures of Dispersion
std_dev_literacy_rate = education_districtwise['OVERALL_LI'].std()
print("Standard Deviation of Literacy Rate:", std_dev_literacy_rate)

variance_literacy_rate = education_districtwise['OVERALL_LI'].var()
print("Variance of Literacy Rate:", variance_literacy_rate)

range_literacy_rate = education_districtwise['OVERALL_LI'].max() - education_districtwise['OVERALL_LI'].min()
print("Range of Literacy Rate:", range_literacy_rate)
```
